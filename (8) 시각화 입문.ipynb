```python
y1 = npr.normal(size = 50).cumsum()
y2 = npr.normal(size = 100).cumsum()
y3 = npr.normal(size = 50).cumsum()
y4 = npr.normal(size = 100).cumsum()

plt.subplot(2, 2, 1)
plt.plot(y1)

# same as plt.subplot(222)
plt.subplot(2, 2, 2)
plt.plot(y2)

# same as plt.subplot(2,2,3)
plt.subplot(223)
plt.plot(y3)

plt.subplot(224)
plt.plot(y4)

plt.tight_layout()
plt.show()
```

```python
# 차트 범위 지정
# 기본적으로 차트에 그려질 x, y 범위에 따라 자동으로 차트의 스케일이 정해지는데, xlim(), ylim()을 이용하여 스케일을 원하는 범위로 설정할 수 있습니다.

# xlim() : x축 범위 설정
# # ylim() : y축 범위 설정
```

```python
y1 = npr.normal(size = 50).cumsum()

# (왼쪽) default xlim, ylim
plt.subplot(121)
plt.plot(y1)
plt.title('Y1')

# (오른쪽) xlim, ylim 사용자 설정
plt.subplot(122)
plt.plot(y1)
plt.xlim(0, 100)
plt.ylim(-10, 10)
plt.title('Y2')
```

```python
## Axes3D
Axes3D 패키지를 이용하여 axes를 3차원 공간으로 생성할 수도 있습니다.

자세한 내용은 Axes3D를 확인
```

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, cmap='hot')

ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.set_zlabel('Z values')
plt.title("3D Surface Plot")
plt.show()
```

```python
## 2차원 공간에 scatter()를 사용한 것처럼 Axes3D 공간에 scatter()를 생성한 모습
```

```python
from mpl_toolkits.mplot3d import Axes3D
X = npr.normal(size = 100)
Y = npr.normal(size = 100)
Z = npr.normal(size = 100)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, marker='o')

ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.set_zlabel('Z values')
plt.title('3D Scatter chart')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 6]
%matplotlib inline
```

```python
# loading 'tips' dataset

tips = sns.load_dataset('tips')
tips.shape
(244, 7)
tips.head()
```

```python
tips.groupby(['sex', 'day']).size()
```

```python
# Basic box plot

plt.boxplot(tips['tip'])

plt.show()
```

```python
# setting outlier symbol, title, xlabel

plt.boxplot(tips['tip'], sym="bo")

plt.title('Box plot of tip')

plt.xticks([1], ['tip'])

plt.show()
```

```python
# Horizontal Box plot with notched box & red color outliers

plt.boxplot(tips['tip'], 

            notch=1, # if 'True' then notched box plot

            sym='rs', # symbol: red square

            vert=0 # vertical : if 'False' then horizontal box plot

           )

plt.show()
```

```python
# Multiple box plots on one Axes

fig, ax = plt.subplots()

ax.boxplot([tips['total_bill'], tips['tip']], sym="b*")

plt.title('Multiple box plots of tips on one Axes')

plt.xticks([1, 2], 

           ['total_bill', 'tip'])

plt.show()
```

```python
## seaborn 으로 여러개 그룹에 대한 상자 그림 그리기¶
  (Grouped box plots by seaborn)
```

```python

# Grouped boxplots by seaborn

import seaborn as sns



sns.boxplot(x="day", 

            y="tip", 

            data=tips)

plt.show()
```

```python
# 위의 요일(day: Thur, Fri, Sat, Sun) 그룹에 더해 hue 옵션을 사용하여 성별(sex: Male, Female)에 따른 그룹을 추가하여 
#(총 4개 그룹 x 2개 그룹 = 8개 그룹) 별 팁(tip) 에 대한 상자 그림을 그려서 비교.
```

```python

#  2 by 4 multi-Grouped boxplots by seaborn

sns.boxplot(x="day", 

            y="tip", 

            hue="sex",

            data=tips)



plt.show()
```

```python
## pandas 로 상자 그림 그리기¶
```

```python
# pandas의 DataFrame에 df.boxplot() 함수를 사용하여 상자 그림 
#(2)번에서 그렸던 그룹 별 상자 그림을 pandas 의 boxplot() 함수로 
```

```python
# Grouped boxplots by pandas

tips.boxplot(column=["tip"], by=["day", "sex"])

plt.show()
```
