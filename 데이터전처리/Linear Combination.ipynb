# 1. ÏÑ†Ìòï Í≤∞Ìï©

ÏÑ†Ìòï Í≤∞Ìï©ÏùÄ ÏàòÌïô, ÌÜµÍ≥ÑÌïô, Î®∏Ïã†Îü¨ÎãùÏóêÏÑú Îß§Ïö∞ Ï§ëÏöîÌïú Í∞úÎÖêÏûÖÎãàÎã§. 

ÏÑ†Ìòï Í≤∞Ìï©ÏùÄ Ïó¨Îü¨ Í∞úÏùò Î≥ÄÏàòÏôÄ Í∞ÄÏ§ëÏπòÍ∞Ä Í≥±Ìï¥ÏßÑ ÌõÑ, Ïù¥Î•º Ìï©ÏÇ∞Ìïú Í∞íÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§.

#### Î≤°ÌÑ∞ÏôÄ ÏÑ†ÌòïÍ≤∞Ìï©

```python
import numpy as np

# ÏûÖÎ†• Î≤°ÌÑ∞ (x)ÏôÄ Í∞ÄÏ§ëÏπò Î≤°ÌÑ∞ (w)
x = np.array([1, 2, 3])
w = np.array([0.5, 1.0, 1.5])

# ÏÑ†Ìòï Í≤∞Ìï© Í≥ÑÏÇ∞
linear_combination = np.dot(w, x)
print("ÏÑ†Ìòï Í≤∞Ìï© Í≤∞Í≥º:", linear_combination)
```

```python
# Ïó¨Í∏∞ÏÑú 
# ùë•=[1,2,3]Ïù¥Í≥†, ùë§=[0.5,1.0,1.5]ÏûÖÎãàÎã§.

# ÏÑ†Ìòï Í≤∞Ìï©ÏùÄ 0.5√ó1+1.0√ó2+1.5√ó3=7.0ÏúºÎ°ú Í≥ÑÏÇ∞Îê©ÎãàÎã§.
```

# ÏÑ†Ìòï ÌöåÍ∑ÄÏóêÏÑúÏùò ÏÑ†Ìòï Í≤∞Ìï©

##### ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏ÏóêÏÑúÎäî ÏûÖÎ†• Î≥ÄÏàòÎì§Í≥º Í∞ÄÏ§ëÏπò Í∞ÑÏùò ÏÑ†Ìòï Í≤∞Ìï©ÏúºÎ°ú Ï∂úÎ†•Í∞íÏùÑ ÏòàÏ∏°Ìï©ÎãàÎã§.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([3, 5, 7, 9, 11])  # y = 2x + 1

# ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏ ÌõàÎ†®
model = LinearRegression()
model.fit(X, y)

# Í∞ÄÏ§ëÏπòÏôÄ Ï†àÌé∏ ÌôïÏù∏
print("Í∞ÄÏ§ëÏπò (w):", model.coef_)
print("Ï†àÌé∏ (b):", model.intercept_)

# ÏòàÏ∏°
y_pred = model.predict(X)
print("ÏòàÏ∏° Í∞í:", y_pred)
```

### ÏÑ†Ìòï Î™®Îç∏ ÏòàÏãú

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = 2 * X.flatten() + 1  # y = 2x + 1

# ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏ ÌõàÎ†®
model = LinearRegression()
model.fit(X, y)

# ÏòàÏ∏°
y_pred = model.predict(X)

# ÏãúÍ∞ÅÌôî
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred, color='red', label='Linear Fit')
plt.legend()
plt.title('Linear Model')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Îã§Ìï≠ ÌöåÍ∑Ä (ÎπÑÏÑ†Ìòï Î™®Îç∏)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Î™®Îç∏ ÌõàÎ†®
model_poly = LinearRegression()
model_poly.fit(X_poly, y)

# ÏòàÏ∏°
y_pred_poly = model_poly.predict(X_poly)

# ÏãúÍ∞ÅÌôî
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred_poly, color='red', label='Polynomial Fit')
plt.legend()
plt.title('Non-linear Model (Polynomial Regression)')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

## Îã§Ìï≠ ÌöåÍ∑Ä

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 5, 10, 17, 26])

# 2Ï∞® Îã§Ìï≠ ÌöåÍ∑Ä Î™®Îç∏
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏ Ï†ÅÏö©
model = LinearRegression()
model.fit(X_poly, y)
y_pred = model.predict(X_poly)

# ÏãúÍ∞ÅÌôî
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred, color='red', label='Polynomial Fit (Degree 2)')
plt.title('Polynomial Regression (Non-linear Relationship)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

# 2. ÏÑ†Ìòï Î≥ÄÌôò (Linear Transformation)

Ï†ïÏùò

ÏÑ†Ìòï Î≥ÄÌôòÏùÄ Î≤°ÌÑ∞ Í≥µÍ∞ÑÏóêÏÑú Î≤°ÌÑ∞Î•º Îã§Î•∏ Î≤°ÌÑ∞Î°ú Î≥ÄÌôòÌïòÎäî Ìï®ÏàòÏûÖÎãàÎã§.

Ïù¥ Î≥ÄÌôòÏùÄ ÏÑ†ÌòïÏÑ±(linearity)ÏùÑ ÎßåÏ°±Ìï¥Ïïº Ìï©ÎãàÎã§

### 1) ÌöåÏ†Ñ(rotation)

ÏΩîÎìú ÏÑ§Î™Ö

```python
# ÏõêÎûò Î≤°ÌÑ∞Îì§
# vectors: ÏõêÎûòÏùò 2Ï∞®Ïõê Î≤°ÌÑ∞Îì§ÏùÑ Ï†ïÏùòÌï©ÎãàÎã§.
# Ï≤´ Î≤àÏß∏ Î≤°ÌÑ∞Îäî [2, 1]ÏûÖÎãàÎã§.
# Îëê Î≤àÏß∏ Î≤°ÌÑ∞Îäî [1, 3]ÏûÖÎãàÎã§.
# Î≤°ÌÑ∞Î•º numpy Î∞∞Ïó¥Î°ú ÏÉùÏÑ±ÌïòÏó¨ ÏÇ¨Ïö©
```

```python
# Î≥ÄÌôò ÌñâÎ†¨
# rotation_matrix: 90ÎèÑ ÌöåÏ†Ñ ÌñâÎ†¨ÏùÑ Ï†ïÏùòÌï©ÎãàÎã§.
# Ïù¥ ÌñâÎ†¨ÏùÄ Î≤°ÌÑ∞Î•º ÏãúÍ≥Ñ Î∞òÎåÄ Î∞©Ìñ•ÏúºÎ°ú 90ÎèÑ ÌöåÏ†ÑÏãúÌÇ§Îäî Ïó≠Ìï†ÏùÑ Ìï©ÎãàÎã§.
```

```python
# ÏÑ†Ìòï Î≥ÄÌôò Ï†ÅÏö©
# vectors.dot(rotation_matrix.T): ÏõêÎûòÏùò Î≤°ÌÑ∞Îì§Ïóê Î≥ÄÌôò ÌñâÎ†¨ÏùÑ Í≥±Ìï¥Ï§çÎãàÎã§.
# rotation_matrix.T: Ï†ÑÏπò ÌñâÎ†¨ÏùÑ ÏÇ¨Ïö©Ìï¥ ÌñâÎ†¨ Í≥±ÏÖàÏùÑ ÏàòÌñâÌï©ÎãàÎã§.
# numpyÏùò .dot() Î©îÏÑúÎìúÎ•º ÏÇ¨Ïö©ÌïòÏó¨ ÌñâÎ†¨ Í≥±ÏÖàÏùÑ ÏàòÌñâÌï©ÎãàÎã§.
# Í≤∞Í≥ºÏ†ÅÏúºÎ°ú Î≥ÄÌôòÎêú Î≤°ÌÑ∞Í∞Ä ÏÉùÏÑ±Îê©ÎãàÎã§. Ïù¥ Î≤°ÌÑ∞Îäî ÏõêÎûò Î≤°ÌÑ∞Î•º ÌöåÏ†Ñ ÌñâÎ†¨ÏùÑ ÏÇ¨Ïö©Ìï¥ Î≥ÄÌôòÌïú Í≤∞Í≥ºÏûÖÎãàÎã§.
```

```python
# ÏõêÎûò Î≤°ÌÑ∞ÏôÄ Î≥ÄÌôòÎêú Î≤°ÌÑ∞ ÏãúÍ∞ÅÌôî
# plt.figure(): ÏÉàÎ°úÏö¥ Í∑∏Î¶ºÏùÑ ÏÉùÏÑ±Ìï©ÎãàÎã§.
# plt.quiver(): Î≤°ÌÑ∞Î•º ÏãúÍ∞ÅÌôîÌïòÍ∏∞ ÏúÑÌï¥ quiver plotÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§.
# [0, 0]: Îëê Î≤°ÌÑ∞Ïùò ÏãúÏûëÏ†êÏùÑ ÏõêÏ†ê (0, 0)ÏúºÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
# vectors[:, 0], vectors[:, 1]: Í∞ÅÍ∞Å xÏ¢åÌëúÏôÄ yÏ¢åÌëúÎ•º ÏßÄÏ†ïÌï©ÎãàÎã§.
# vectors[:, 0]ÏùÄ Î™®Îì† Î≤°ÌÑ∞Ïùò x Ï¢åÌëú ([2, 1]).
# vectors[:, 1]ÏùÄ Î™®Îì† Î≤°ÌÑ∞Ïùò y Ï¢åÌëú ([1, 3]).
# angles='xy', scale_units='xy', scale=1: Î≤°ÌÑ∞Ïùò Ïä§ÏºÄÏùº Î∞è Î∞©Ìñ•ÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
# color='blue': ÏõêÎûò Î≤°ÌÑ∞Ïùò ÏÉâÏÉÅÏùÑ ÌååÎûÄÏÉâÏúºÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
# label='Original Vectors': ÏõêÎûò Î≤°ÌÑ∞Ïóê Î†àÏù¥Î∏îÏùÑ ÏßÄÏ†ïÌï©ÎãàÎã§.
# Îëê Î≤àÏß∏ plt.quiver():
# Î≥ÄÌôòÎêú Î≤°ÌÑ∞Î•º ÏãúÍ∞ÅÌôîÌïòÎäî ÏΩîÎìúÏûÖÎãàÎã§.
# transformed_vectors[:, 0], transformed_vectors[:, 1]Îäî Î≥ÄÌôòÎêú Î≤°ÌÑ∞Îì§Ïùò xÏ¢åÌëú Î∞è yÏ¢åÌëúÏûÖÎãàÎã§.
# ÏÉâÏÉÅÏùÄ Îπ®Í∞ÑÏÉâÏúºÎ°ú ÏßÄÏ†ïÌïòÍ≥†, Î†àÏù¥Î∏îÏùÑ 'Transformed Vectors'Î°ú ÏÑ§Ï†ïÌï©ÎãàÎã§.
```

```python
# Î≤°ÌÑ∞ ÏãúÍ∞ÅÌôî
# 1,plt.xlim(-4, 4) :xÏ∂ïÏùò Î≤îÏúÑÎ•º ÏÑ§Ï†ïÌï©ÎãàÎã§.
# 2. plt.ylim(-4, 4) : yÏ∂ïÏùò Î≤îÏúÑÎ•º ÏÑ§Ï†ïÌï©ÎãàÎã§.
# 3. plt.axhline(0, color='black', lw=0.5) : ÏàòÌèâÏÑ†(horizontal line)ÏùÑ Í∑∏Î¶ΩÎãàÎã§.
# y = 0Ïù∏ ÏúÑÏπòÏóê Í≤ÄÏ†ïÏÉâÏùò ÏàòÌèâÏÑ†ÏùÑ Í∑∏Î¶ΩÎãàÎã§.
# lw=0.5Îäî ÏÑ†Ïùò ÎëêÍªò(linewidth)Î•º ÏùòÎØ∏ÌïòÎ©∞, 0.5Î°ú ÏÑ§Ï†ïÌïòÏó¨ ÏñáÏùÄ ÏÑ†ÏùÑ Í∑∏Î¶ΩÎãàÎã§.
# Ïù¥ ÏàòÌèâÏÑ†ÏùÄ xÏ∂ïÏùÑ ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú ÌëúÏãúÌïòÏó¨ Í∑∏ÎûòÌîÑ Ìï¥ÏÑùÏùÑ ÏâΩÍ≤å ÎßåÎì≠ÎãàÎã§.
# 4. plt.axvline(0, color='black', lw=0.5) : ÏàòÏßÅÏÑ†(vertical line)ÏùÑ Í∑∏Î¶ΩÎãàÎã§.
# x = 0Ïù∏ ÏúÑÏπòÏóê Í≤ÄÏ†ïÏÉâÏùò ÏàòÏßÅÏÑ†ÏùÑ Í∑∏Î¶ΩÎãàÎã§.
# Ïù¥ ÏàòÏßÅÏÑ†ÏùÄ yÏ∂ïÏùÑ ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú ÌëúÏãúÌïòÏó¨ Í∑∏ÎûòÌîÑÏùò Ï§ëÏã¨ÏùÑ Î™ÖÌôïÌïòÍ≤å Î≥¥Ïó¨Ï§çÎãàÎã§.
# 5. plt.grid() : Í∑∏ÎûòÌîÑÏóê Í∑∏Î¶¨Îìú(grid)Î•º Ï∂îÍ∞ÄÌï©ÎãàÎã§.
# 6. plt.gca().set_aspect('equal') : Í∑∏ÎûòÌîÑÏùò Ï∂ï ÎπÑÏú®ÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
# gca()**Îäî ÌòÑÏû¨Ïùò Í∑∏ÎûòÌîÑ Í∞ùÏ≤¥Î•º Î∞òÌôòÌïòÎäî Ìï®ÏàòÏù¥Î©∞, **.set_aspect('equal')**ÏùÄ xÏ∂ïÍ≥º yÏ∂ïÏùò Îã®ÏúÑÎ•º ÎèôÏùºÌïòÍ≤å ÏÑ§Ï†ïÌï©ÎãàÎã§.
# Ïù¥ ÏÑ§Ï†ïÏùÄ Î≤°ÌÑ∞Í∞Ä Í∏∞ÌïòÌïôÏ†ÅÏúºÎ°ú Ï†ïÌôïÌïòÍ≤å ÌëúÌòÑÎêòÎèÑÎ°ù ÌïòÍ∏∞ ÏúÑÌï¥ Ï§ëÏöîÌï©ÎãàÎã§. 
# ÏòàÎ•º Îì§Ïñ¥, 1Îã®ÏúÑÍ∞Ä xÏ∂ïÍ≥º yÏ∂ïÏóêÏÑú ÎèôÏùºÌïú Í∏∏Ïù¥Î°ú Î≥¥Ïù¥ÎèÑÎ°ù ÏÑ§Ï†ïÌïòÏó¨ ÏôúÍ≥° ÏóÜÏù¥ Î≤°ÌÑ∞Ïùò ÌÅ¨Í∏∞ÏôÄ Î∞©Ìñ•ÏùÑ Ïù¥Ìï¥Ìï† Ïàò ÏûàÍ≤å Ìï©ÎãàÎã§.
# 7. plt.legend() : Î≤îÎ°Ä(legend)Î•º Ï∂îÍ∞ÄÌï©ÎãàÎã§.
# plt.quiver()ÏóêÏÑú ÏßÄÏ†ïÌïú label Í∞í ('Original Vectors'ÏôÄ 'Transformed Vectors')Ïù¥ Î≤îÎ°ÄÏóê ÌëúÏãúÎêòÏñ¥, Ïñ¥Îñ§ ÌôîÏÇ¥ÌëúÍ∞Ä ÏõêÎûò Î≤°ÌÑ∞Ïù∏ÏßÄ, Ïñ¥Îñ§ ÌôîÏÇ¥ÌëúÍ∞Ä Î≥ÄÌôòÎêú Î≤°ÌÑ∞Ïù∏ÏßÄÎ•º Ïïå Ïàò ÏûàÏäµÎãàÎã§.
# 8. plt.show() : Í∑∏ÎûòÌîÑÎ•º ÌôîÎ©¥Ïóê ÌëúÏãúÌï©ÎãàÎã§.
# Ïù¥ Î™ÖÎ†πÏùÑ Ïã§ÌñâÌï¥Ïïº Í∑∏ÎûòÌîÑÍ∞Ä Ïã§Ï†úÎ°ú Î≥¥Ïù¥Í≤å Îê©ÎãàÎã§.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ÏõêÎûò Î≤°ÌÑ∞Îì§
vectors = np.array([[2, 1], [1, 3]])

# Î≥ÄÌôò ÌñâÎ†¨ (90ÎèÑ ÌöåÏ†Ñ ÌñâÎ†¨)
rotation_matrix = np.array([[0, -1], [1, 0]])

# ÏÑ†Ìòï Î≥ÄÌôò Ï†ÅÏö©
transformed_vectors = vectors.dot(rotation_matrix.T)

# ÏõêÎûò Î≤°ÌÑ∞ÏôÄ Î≥ÄÌôòÎêú Î≤°ÌÑ∞ ÏãúÍ∞ÅÌôî
plt.figure()
plt.quiver([0, 0], [0, 0], vectors[:, 0], vectors[:, 1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vectors')
plt.quiver([0, 0], [0, 0], transformed_vectors[:, 0], transformed_vectors[:, 1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed Vectors')

plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.grid()
plt.gca().set_aspect('equal')
plt.legend()
plt.show()
```

```python
# ÏΩîÎìú ÏÑ§Î™Ö
# Î≤°ÌÑ∞ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±:

# vectorsÎäî 2D Í≥µÍ∞ÑÏùò ÏûÖÎ†• Î≤°ÌÑ∞Îì§ÏûÖÎãàÎã§.
# originÏùÄ Î≤°ÌÑ∞Îì§Ïù¥ ÏãúÏûëÌïòÎäî ÏõêÏ†êÏûÖÎãàÎã§.
# Î≥ÄÌôò ÌñâÎ†¨ Ï†ïÏùò:

# transformation_matrixÎäî Î≤°ÌÑ∞Î•º Î≥ÄÌôòÌïòÎäî Îç∞ ÏÇ¨Ïö©ÎêòÎäî ÌñâÎ†¨ÏûÖÎãàÎã§.
# Ïù¥ ÏòàÏãúÏóêÏÑúÎäî ÌñâÎ†¨ Î•º ÏÇ¨Ïö©ÌïòÏó¨ Î≤°ÌÑ∞Î•º ÌôïÎåÄÌïòÍ≥† ÌöåÏ†ÑÏãúÌÇµÎãàÎã§.

#     ÏÑ†Ìòï Î≥ÄÌôò Ï†ÅÏö©:

# @ Ïó∞ÏÇ∞ÏûêÎäî ÌñâÎ†¨ Í≥±ÏÖàÏùÑ ÏàòÌñâÌï©ÎãàÎã§.
# vectors @ transformation_matrix.TÎäî Í∞Å Î≤°ÌÑ∞Ïóê Î≥ÄÌôò ÌñâÎ†¨ÏùÑ Ï†ÅÏö©Ìïú Í≤∞Í≥ºÏûÖÎãàÎã§.
# ÏãúÍ∞ÅÌôî:

# quiver Ìï®ÏàòÎäî 2D Î≤°ÌÑ∞Î•º ÏãúÍ∞ÅÌôîÌï©ÎãàÎã§.
# ÌååÎûÄÏÉâ Î≤°ÌÑ∞Îäî Î≥ÄÌôò Ï†ÑÏùò ÏõêÎ≥∏ Î≤°ÌÑ∞, Îπ®Í∞ÑÏÉâ Î≤°ÌÑ∞Îäî Î≥ÄÌôò ÌõÑÏùò Î≤°ÌÑ∞ÏûÖÎãàÎã§.
# Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï:

# Ï∂ïÍ≥º Í≤©ÏûêÎ•º Ï∂îÍ∞ÄÌïòÍ≥†, Í∑∏ÎûòÌîÑÏùò Ï†úÎ™©Í≥º Î†àÏù¥Î∏îÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
# Ï∂úÎ†• Í≤∞Í≥º
# Í∑∏ÎûòÌîÑÏóêÎäî **ÏõêÎ≥∏ Î≤°ÌÑ∞(ÌååÎûÄÏÉâ)**ÏôÄ **Î≥ÄÌôòÎêú Î≤°ÌÑ∞(Îπ®Í∞ÑÏÉâ)**Í∞Ä ÌëúÏãúÎê©ÎãàÎã§.
# Î≥ÄÌôòÎêú Î≤°ÌÑ∞Îì§ÏùÄ ÌôïÎåÄÎêòÍ≥† ÌöåÏ†ÑÎêú ÌòïÌÉúÎ°ú ÎÇòÌÉÄÎÇ©ÎãàÎã§.
# Ïù¥ ÏΩîÎìúÎäî ÏÑ†Ìòï Î≥ÄÌôòÏù¥ Î≤°ÌÑ∞Ïùò ÌÅ¨Í∏∞ÏôÄ Î∞©Ìñ•ÏùÑ Ïñ¥ÎñªÍ≤å Î≥ÄÍ≤ΩÌïòÎäîÏßÄ ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Î≥¥Ïó¨Ï§çÎãàÎã§.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 2D Î≤°ÌÑ∞ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
vectors = np.array([[1, 2], [2, 3], [-1, -1], [-2, -2]])
origin = np.zeros(vectors.shape)  # ÏõêÏ†ê (0, 0)

# 2. Î≥ÄÌôò ÌñâÎ†¨ Ï†ïÏùò (ÌôïÎåÄÏôÄ ÌöåÏ†ÑÏùÑ ÎÇòÌÉÄÎÉÑ)
transformation_matrix = np.array([[2, 1], [-1, 1]])

# 3. ÏÑ†Ìòï Î≥ÄÌôò Ï†ÅÏö©
transformed_vectors = vectors @ transformation_matrix.T

# 4. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 6))

# ÏõêÎ≥∏ Î≤°ÌÑ∞ ÏãúÍ∞ÅÌôî (ÌååÎûÄÏÉâ)
plt.quiver(*origin.T, vectors[:, 0], vectors[:, 1], color='blue', scale=5, label='Original Vectors')

# Î≥ÄÌôòÎêú Î≤°ÌÑ∞ ÏãúÍ∞ÅÌôî (Îπ®Í∞ÑÏÉâ)
plt.quiver(*origin.T, transformed_vectors[:, 0], transformed_vectors[:, 1], color='red', scale=5, label='Transformed Vectors')

# 5. Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid()
plt.legend()
plt.title('Linear Transformation of Vectors')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.gca().set_aspect('equal')

# 6. Í∑∏ÎûòÌîÑ Ï∂úÎ†•
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1, 4, 9, 16, 25])  # y = x^2

# Îã§Ìï≠ ÌäπÏÑ± ÏÉùÏÑ± (2Ï∞® Îã§Ìï≠Ïãù)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Îã§Ìï≠ ÌöåÍ∑Ä Î™®Îç∏ ÌõàÎ†®
model = LinearRegression()
model.fit(X_poly, y)

# ÏòàÏ∏°
y_pred = model.predict(X_poly)

# ÏãúÍ∞ÅÌôî
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred, color='red', label='Polynomial Fit (Degree 2)')
plt.legend()
plt.title('Polynomial Regression Example')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
