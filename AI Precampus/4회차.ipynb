#### 2.1. 리스트(List)
- 리스트: 순서가 있는 변경 가능한 시퀀스 데이터 타입을 말한다.
- 파이썬에서 변수가 어떤 값을 바인딩(binding)하고 있는 경우
- 바인딩(binding) : 프로그램에 사용된 구성 요소의 실제 값 또는 프로퍼티를 결정짓는 행위를 의미한다.
- 프롬프트 상에서 변수명을 입력한 후 엔터 키를 누르면 해당 변수가 가리키는 값이 반환한다. 

- 파이썬에서는 리스트(list) 이용하면 여러 데이터를 한 번에 저장 가능--> 변수명을 여러 개 사용할 필요가 없다.
- 여러 데이터를 리스트에 담으려면 대괄호'[]' 사이에 데이터를 쉼표로 구분해서 넣으면 됨.
- 리스트 : 대괄호'[] 로 요소를 기입한다. 튜플 : 소괄호() , 디셔너리 : 중괄호{}
- 요소(element) : list 를 이루는 값을 말한다.
- 인덱스(index) : 오소들의 순서에 따른 번호 --> 시작은 0 부터임을 주의할 것

- 리스트는 대괄호'[]' 사용, 튜플은 소괄호'()' 사용에 인자를 담아 연속형 데이터를 처리하는데 사용

#### 2. 3 딕셔너리(dictionary)

- 딕셔너리 : 키(key)와 값(value)이라는 것을 쌍으로 저장함으로써 더 쉽게 저장된 값을 찾을 수 있는 구조를 말한다.
- key : 중복 값을 사용할 수 없고, 변경이 불가하다.
- value : key에 해당하는 값으로 수정이 가능하다.
- 딕셔너리는 중괄호 '{}' 기호 사용 : 리스트 대괄호[] , 튜플 소괄호()
- 요소가 여러개인 경우 콤마(,)로 구분
- 딕셔너리는 리스트와 튜플과 달리 인덱싱을 지원 않음.
- 사례 : 범죄종류별 나이, 범죄별 소득수준

```python
city
```

```python
# 직접 입력하기
# key : 서울 value : 500.......
# 반드시 딕셔너리 구조를 위해 중괄호로 해야 한다.

city_sales = {'서울':500, '부산':350, '대구':200}
```

### 3. 라이브러리 (Library)

- 파이썬 라이브러리(Python Library)는 특정 기능을 쉽게 구현할 수 있도록 미리 작성된 파이썬 코드들의 집합 
- 쉽게 말해, 프로그래밍에서 자주 사용하는 기능을 묶어놓은 도구 상자라고 생각하면 됩니다.

- 1. 라이브러리의 역할
- 코드 재사용: 매번 직접 코드를 작성하지 않고, 필요한 기능을 불러와서 바로 사용 가능.
- 개발 효율성 향상: 데이터 분석, 웹 개발, 인공지능 등 다양한 분야의 복잡한 기능을 손쉽게 구현
- 표준화된 기능 제공: 여러 개발자가 공통적으로 사용하는 검증된 기능을 제공하여 안정성과 신뢰성을 높임

- 2. 라이브러리의 종류
- 파이썬 라이브러리는 크게 두 가지로 나눌 수 있습니다.
- (1) 표준 라이브러리 (Standard Library)
- 파이썬을 설치하면 자동으로 제공되는 기본 라이브러리
- 예를 들어:
- math → 수학 계산 함수 제공 (sqrt, sin, log 등)
- datetime → 날짜와 시간 처리
- os → 운영체제와 상호작용 (파일 경로, 디렉터리 관리 등)
- json → JSON 데이터 읽기/쓰기

- (2) 외부 라이브러리 (Third-party Library)
- 파이썬 기본 설치에는 포함되지 않지만, pip 명령어를 사용해 설치 후 활용하는 라이브러리

##### 자료구조 : Series와 DataFrame¶

- Series: 1차원 벡터 형식 , 하나의 열이 되는 데이터를 리스트나 일차원 배열을 준비한다.
- DataFrame: 2차원 행렬 형식 행(row)과 열(column)로 구성
- 행(row) : 가로줄 , 열(column) : 세로줄

- Series : 1차원 형태의 자료구조
- 파이썬의 리스트(list)와 비슷하고 , 딕셔너리와 유사한 자료구조
- Series모든 데이터 유형(정수, 문자열, 부동 소수점 숫자, Python 객체 등)을 보유할 수 있는 1차원 레이블 배열
- Series형태의 자료구조는 기본적으로 index와 value 가 존재하기 때문에 일반적으로 생각하는 시계열 형태와 일치

```python
# 라이브러리 불러오기

import pandas as pd

income_list = pd.Series([45652, 26392, 46405],index=["남성","여성","여성"])
income_list
```

```python
# 인덱스 이름 지정
# pd.Series()함수의 옵션으로 index명을 지정

income_list.index.name = "성별","총소득"
income_list
```

##### 1. DataFrame 생성하기

```python
# 기본 방법 3가지 ( 수집된 데이터가 엑셀등의 자료가 아닌 경우)
```

```python
# 첫번째 방법
# 2차원 표에서 column 단위로 데이터를 표현하여 dataframe 생성
# ==> column명을 딕셔너리의 key로 (아래에서 sales) ,  데이터는 딕셔너리의 value로
```

```python
from pandas import DataFrame

sales={"사과":[10,5,20], "배":[55,65,40], "귤":[150,200,350]}

city =["서울", "부산", "인천"]

df=DataFrame(sales, city)
print(df)
```

```python
# 두번째 방법
# 2차원 표에서 row 단위로 데이터를 리스트로 표현하여 dataframe 생성
# data에 해당하는 아래 sales, index에 해당하는 city , columns 을 모두 리스트로 표현하여 생성하는 방법
```

```python
from pandas import DataFrame

sales=[[10,55,150], [5,65,200], [20,40,350]]

city =["서울", "부산", "인천"]
columns=["사과","배","귤"]

df1=DataFrame(data=sales, index=city,columns=columns)
print(df1)
```

```python
# 세번째 방법
# 2차원 표에서 row 단위로 데이터를 딕셔너리로 표현하여 dataframe 생성
```

```python
from pandas import DataFrame

sales=[{"사과":10, "배":55, "귤":150},
       {"사과":5, "배":65, "귤":200},
       {"사과":20, "배":40, "귤":350}]

city =["서울", "부산", "인천"]

df2=DataFrame(data= sales, index=city)
print(df2)
```
