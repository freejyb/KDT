# 2. 바이올린 플롯 (Violin Plot)

바이올린 플롯(Violin Plot)은 박스플롯(Boxplot)과 커널 밀도 추정(KDE: Kernel Density Estimation)**을 결합한 그래프로, 

데이터의 분포를 시각화하는 데 사용

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 랜덤 데이터 생성
np.random.seed(42)
data = {
    'Category': ['A'] * 100 + ['B'] * 100 + ['C'] * 100,
    'Values': np.concatenate([np.random.normal(0, 1, 100), np.random.normal(1, 1.5, 100), np.random.normal(2, 0.5, 100)])
}
df = pd.DataFrame(data)

# 바이올린 플롯 생성
plt.figure(figsize=(10, 6))
sns.violinplot(x='Category', y='Values', data=df, inner='quartile', palette='pastel')
plt.title('Violin Plot Example')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

### 바이올린 플롯과 박스플롯 비교

```python
plt.figure(figsize=(12, 6))

# 박스플롯 생성
plt.subplot(1, 2, 1)
sns.boxplot(x='Category', y='Values', data=df, palette='pastel')
plt.title('Boxplot')

# 바이올린 플롯 생성
plt.subplot(1, 2, 2)
sns.violinplot(x='Category', y='Values', data=df, inner='quartile', palette='pastel')
plt.title('Violin Plot')

plt.show()
```

# 3. 바 플롯 (Barplot)

바 플롯(Barplot)은 범주형 데이터의 값을 비교하고, 각 범주(category) 간의 차이를 시각적으로 나타내는 그래프

### 1) Matplotlib 예제 (Vertical Bar Plot) 

```python
import matplotlib.pyplot as plt

# 데이터 생성
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# 바 플롯 생성
plt.bar(categories, values, color='skyblue')
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

# 4. 히스토그램

스토그램(Histogram)은 연속형 데이터의 분포를 나타내는 그래프로, 데이터를 여러 구간(bin)으로 나누고, 각 구간에 속하는 데이터의 빈도를 시각화합니다. 

히스토그램은 데이터의 밀도, 분포, 변동성 등을 이해하는 데 유용하며, 탐색적 데이터 분석(EDA)에서 자주 사용

### 1) Matplotlib 예제

```python
import matplotlib.pyplot as plt
import numpy as np

# 랜덤 데이터 생성
np.random.seed(42)
data = np.random.randn(1000)

# 히스토그램 생성
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

### 2) Seaborn 예제

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 랜덤 데이터 생성
data = np.random.randn(1000)

# 히스토그램 생성
sns.histplot(data, kde=True, color='lightgreen')
plt.title('Histogram with KDE')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# 5. 라인 플롯 (Line Plot)

라인 플롯(Line Plot)**은 데이터를 선으로 연결하여 시각화하는 그래프 

라인 플롯은 주로 시간의 흐름에 따른 변화를 시각화하거나, 연속적인 데이터의 패턴을 파악하는 데 사용 

특히, 시계열 데이터(time series data)를 분석할 때 자주 사용

### 1) Matplotlib 예제

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 라인 플롯 생성
plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='red')
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

### 2) Seaborn 예제

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 생성
data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales_A': [150, 180, 200, 220, 210, 250],
    'Sales_B': [130, 160, 180, 200, 195, 230]
})

# 데이터 변환 (Seaborn에서는 길게 변환된 데이터 사용)
df = pd.melt(data, id_vars=['Month'], value_vars=['Sales_A', 'Sales_B'], var_name='Product', value_name='Sales')

# 라인 플롯 생성
sns.lineplot(x='Month', y='Sales', hue='Product', data=df)
plt.title('Monthly Sales Comparison')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

월별 판매량 데이터를 사용하여 두 제품의 판매 추이를 비교

```python
# 코드 설명
# 데이터 생성:

# months: 12개월의 이름을 나타내는 리스트입니다.
# sales_A, sales_B: 두 제품의 월별 판매량 데이터를 나타냅니다.
# 마커가 추가된 라인 플롯:

# Product A:
# marker='o': 원(circle) 모양의 마커를 사용합니다.
# color='blue', markerfacecolor='lightblue': 선 색상은 파란색, 마커의 내부 색상은 연한 파란색입니다.
# markersize=8: 마커 크기를 8로 설정합니다.
# markeredgecolor='black': 마커의 테두리 색상은 검은색입니다.
# Product B:
# marker='s': 사각형(square) 모양의 마커를 사용합니다.
# color='red', markerfacecolor='lightcoral': 선 색상은 빨간색, 마커의 내부 색상은 연한 빨간색입니다.
# 그래프 꾸미기:

# title(), xlabel(), ylabel()로 그래프 제목과 축 레이블을 설정합니다.
# legend()는 범례를 추가합니다.
# grid(True)는 그래프에 격자를 추가하여 가독성을 높입니다.
```

# 6. 산점도 (Scatter Plot)

산점도(Scatter Plot)는 두 변수 간의 관계를 시각화하는 그래프

데이터 포인트를 X축과 Y축에 표시하여 각 점이 나타내는 값을 보여줍니다. 

산점도는 두 변수 간의 상관관계(correlation)를 분석하고, 데이터의 패턴, 분포, 이상치(outliers)를 파악하는 데 매우 유용한 도구

#### 1) Matplotlib 예제

```python
# 코드 설명
# 데이터 생성:
# x는 0에서 10 사이의 난수를 생성합니다.
# y는 y = 2.5 * x의 선형 방정식에 잡음을 더한 값입니다.
# 산점도 생성:
# plt.scatter() 함수를 사용하여 산점도를 그립니다.
# color='blue'는 점의 색상을 파란색으로 설정합니다.
# alpha=0.7은 점의 투명도를 설정합니다.
# edgecolors='black'은 점의 테두리 색상을 검은색으로 설정합니다.
# 그래프 꾸미기:
# title(), xlabel(), ylabel()로 그래프 제목과 축 레이블을 설정합니다.
# grid(True)는 그래프에 격자를 추가하여 가독성을 높입니다.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
np.random.seed(42)
x = np.random.rand(100) * 10  # X축 데이터
y = 2.5 * x + np.random.randn(100) * 5  # Y축 데이터 (y = 2.5x + 잡음)

# 산점도 생성
plt.scatter(x, y, color='blue', alpha=0.7, edgecolors='black')
plt.title('Scatter Plot Example')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.show()
```

### 2) Seaborn 예제

```python
import seaborn as sns
import pandas as pd

# 데이터 생성
df = pd.DataFrame({'X': x, 'Y': y})

# 산점도 생성
sns.scatterplot(x='X', y='Y', data=df, color='red', edgecolor='black')
plt.title('Scatter Plot with Seaborn')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.show()
```

# 7. Regplot (Regression Plot)

Regplot은 Seaborn 라이브러리에서 제공하는 시각화 도구로, 산점도(scatter plot)와 회귀선(regression line)을 함께 그려 데이터의 관계를 분석하는 데 사용됩니다. 

    Regplot은 두 변수 간의 선형 관계를 시각적으로 보여주며, 데이터의 패턴과 추세를 쉽게 파악할 수 있게 합니다.

```python
# 코드 설명
# 데이터 생성:

# x는 0에서 10 사이의 난수를 생성합니다.
# y는 y = 2.5 * x의 선형 방정식에 잡음을 더한 값입니다.
# 두 변수 간의 선형 관계를 관찰할 수 있도록 데이터를 생성합니다.
# Regplot 생성:

# sns.regplot() 함수는 산점도와 회귀선을 함께 그립니다.
# line_kws={'color': 'red'}는 회귀선의 색상을 빨간색으로 설정합니다.
# marker='o'는 산점도의 마커 스타일을 원(circle)으로 설정합니다.
# 그래프 꾸미기:

# title(), xlabel(), ylabel()로 그래프 제목과 축 레이블을 설정합니다.
# grid(True)는 그래프에 격자를 추가합니다.
# 3. Regplot의 주요 옵션
# 옵션	설명
# x	X축 변수
# y	Y축 변수
# data	데이터프레임
# color	점과 회귀선의 색상
# marker	산점도의 마커 스타일
# line_kws	회귀선의 스타일 (색상, 굵기 등)
# ci	오차 구간(confidence interval) 표시
# fit_reg	회귀선 표시 여부 (True/False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1. 데이터 생성
np.random.seed(42)
x = np.random.rand(100) * 10  # 0에서 10 사이의 난수
y = 2.5 * x + np.random.randn(100) * 5  # y = 2.5x + 잡음

# 2. 데이터프레임 생성
df = pd.DataFrame({'X': x, 'Y': y})

# 3. Regplot 생성
plt.figure(figsize=(10, 6))
sns.regplot(x='X', y='Y', data=df, color='blue', marker='o', line_kws={'color': 'red'})
plt.title('Regplot Example')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.show()
```

# 8. Pairplot

페어플롯(Pairplot)은 Seaborn 라이브러리에서 제공하는 시각화 도구

데이터프레임의 모든 변수 쌍(pair) 간의 관계를 한눈에 파악 가능 

페어플롯은 각 변수 간의 산점도(scatter plot)와 히스토그램(histogram) 또는 커널 밀도 추정(KDE: Kernel Density Estimation)을 함께 표시


```python
# 코드 설명
# 데이터 로드:

# sns.load_dataset('iris')는 Seaborn에서 제공하는 Iris 데이터셋을 로드합니다.
# 이 데이터셋에는 꽃받침(sepal)과 꽃잎(petal)의 길이 및 너비, 그리고 품종(species) 정보가 포함되어 있습니다.
# 페어플롯 생성:

# sns.pairplot() 함수는 데이터프레임의 모든 변수 간의 관계를 시각화합니다.
# hue='species'는 품종별로 색상을 다르게 설정합니다.
# palette='Set2'는 색상 팔레트를 지정합니다.
# diag_kind='kde'는 대각선에 히스토그램 대신 **커널 밀도 추정(KDE)**을 표시합니다.
# 그래프 출력:

# plt.show()를 사용하여 그래프를 출력합니다.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 데이터 로드 (Iris 데이터셋 사용)
df = sns.load_dataset('iris')

# 2. 페어플롯 생성
sns.pairplot(df, hue='species', palette='Set2', diag_kind='kde')

# 3. 그래프 출력
plt.show()
```

# 9. 히트 맵(heat map)

히트맵(Heatmap)은 데이터 값을 색상으로 표현한 그래프로, 데이터의 패턴과 상관관계를 시각적으로 나타냅니다. 

특히, 행렬 형태의 데이터나 상관 행렬(correlation matrix)을 시각화할 때 자주 사용됩니다.

    색상의 농도나 명도는 데이터 값의 크기를 나타내며, 한눈에 데이터의 분포나 관계를 파악할 수 있게 도와줍니다.

```python
# 코드 설명
# 데이터 로드:

# sns.load_dataset('iris')는 Seaborn에서 제공하는 Iris 데이터셋을 로드합니다.
# 상관 행렬 계산:

# df.corr()는 변수 간의 상관계수를 계산하여 상관 행렬을 생성합니다.
# 히트맵 생성:

# sns.heatmap() 함수는 히트맵을 생성합니다.
# annot=True는 각 셀에 상관계수 값을 표시합니다.
# cmap='coolwarm'은 색상 맵을 설정합니다.
# linewidths=0.5는 셀 간의 경계선을 나타냅니다.

# 3. 히트맵의 주요 옵션
# 옵션	설명
# data	히트맵에 표시할 데이터 (상관 행렬 등)
# annot	셀 안에 데이터 값을 표시할지 여부 (True/False)
# cmap	색상 맵 (예: 'viridis', 'coolwarm')
# linewidths	셀 간의 경계선 두께
# vmin, vmax	색상의 최소/최대값 설정
# center	색상 맵의 중심값 설정
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 1. 데이터 로드 (Iris 데이터셋 사용)
df = sns.load_dataset('iris')

# 2. 숫자형 데이터만 선택
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# 3. 상관 행렬 계산
corr_matrix = numeric_df.corr()

# 4. 히트맵 생성
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Dataset (Numeric Data Only)')
plt.show()
```

# 10. Pie Chart

파이 차트(Pie Chart)는 데이터를 원(circle) 형태로 시각화하여, 비율이나 퍼센트(%)를 나타내는 그래프


```python
# 코드 설명
# 데이터 정의:
# labels: 파이 차트의 각 조각에 해당하는 레이블입니다.
# sizes: 각 항목의 비율입니다.
# colors: 조각의 색상입니다.
# explode: 특정 조각을 강조하여 원에서 분리할 때 사용합니다.
# 파이 차트 생성:
# plt.pie() 함수를 사용하여 파이 차트를 생성합니다.
# autopct='%1.1f%%'는 조각 위에 백분율을 표시합니다.
# startangle=140은 원의 시작 각도를 설정합니다.
# 그래프 출력:
# plt.show()로 파이 차트를 출력합니다.
```

```python
# 파이 차트의 주요 옵션
# 옵션	설명
# labels	각 조각의 레이블
# sizes	각 조각의 크기 (비율)
# colors	조각의 색상
# explode	특정 조각을 강조 (분리)
# autopct	백분율 표시 형식 (%1.1f%% 등)
# startangle	시작 각도 (기본값: 0도)
# shadow	그림자 효과 추가 (True/False)
```

```python
import matplotlib.pyplot as plt

# 1. 데이터 정의
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 35, 20, 20]
colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightyellow']
explode = [0, 0.1, 0, 0]  # 'explode'는 강조할 조각을 띄웁니다.

# 2. 파이 차트 생성
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Example')
plt.show()
```

도넛 챠트

```python
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=140, wedgeprops={'width': 0.4})
plt.title('Donut Chart Example')
plt.show()
```

```python
# 질적자료(범주형)에 대한 그래프
# 파이 챠트 : 원을 그린후 그 원에 각 계급의 상대도수에 대응하는 면적 또는 부분으로 나눈다.

# 도수(frequency) : 각 범주에 속하는 관측값의 개수 (건수)
# 상대도수(relative frequency) : 도수를 전체 개수로 나눈 비율==> autopct
# 도수분포표(frequency table) : 범주형 자료에서 범주와 그 범주에 대응하는 도수와 상대도수를 나열하는 표

# plt.pie() :  파이 모양의 차트를 표현 가능
# autopct : 각 파이 각각에 문자열을 출력하는 것을 설정
```
