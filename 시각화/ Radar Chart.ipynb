레이더 차트(방사형 그래프, Radar Chart 또는 Spider Chart)

레이더 차트(방사형 그래프, Radar Chart 또는 Spider Chart)는 여러 항목(차원)의 값을 원형 축에 따라 표시하는 시각화 방식

보통 범주별 비교

## 1. 기본 레이더 차트 예제 (matplotlib)

```python
import matplotlib.pyplot as plt
import numpy as np


#wndows에서 한글이 잘 보이는 기본 폰트 설정 (예: 'Malgun Gothic')
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 음수 부호 깨짐 방지


# 항목(label)과 값 정의
labels = ['성장성', '수익성', '안정성', '시장지배력', '배당']
values = [80, 65, 90, 70, 60]

# 각도 계산 및 원형 닫기
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# 값과 라벨 모두 처음 항목을 다시 붙여서 원 닫기
values += [values[0]]
angles += [angles[0]]
labels += [labels[0]]

# 레이더 차트 그리기
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

ax.plot(angles, values, linewidth=2, linestyle='solid', color='teal')
ax.fill(angles, values, alpha=0.25, color='teal')

# 축 설정
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), labels[:-1])  # 마지막 라벨은 제외해서 중복 방지
ax.set_title("재무 항목별 점수 (레이더 차트)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ▶️ 데이터
labels = ['성장성', '수익성', '안정성', '시장지배력', '배당']
samsung = [80, 75, 90, 85, 60]
skhynix = [90, 70, 65, 75, 40]

# ▶️ 원형 닫기용 첫 항목 추가
labels_with_loop = labels + [labels[0]]
samsung += [samsung[0]]
skhynix += [skhynix[0]]

# ▶️ 각 축 각도 설정
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += [angles[0]]  # 닫기용

# ▶️ 그리기
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

ax.plot(angles, samsung, label='삼성전자', color='blue')
ax.fill(angles, samsung, alpha=0.25, color='blue')

ax.plot(angles, skhynix, label='SK하이닉스', color='green')
ax.fill(angles, skhynix, alpha=0.25, color='green')

# ▶️ 설정
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), labels)  # 여기 수정됨
ax.set_title("삼성전자 vs SK하이닉스 재무 특성 비교", fontsize=14)
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

plt.tight_layout()
plt.show()
```
