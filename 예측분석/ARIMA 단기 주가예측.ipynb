## 삼성전자 주가 데이터

```python
# 삼성전자 주가 데이터를 불러옵니다.
# ARIMA 모델을 사용해 데이터를 학습합니다.
# 내일의 주가를 예측합니다.
```

```python
# order=(5, 1, 0)은 ARIMA 모델의 (p, d, q) 값을 의미합니다. 이를 적절히 조정하여 더 나은 예측 결과를 얻을 수 있습니다. 
```

```python
import pandas as pd
from pykrx import stock
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# 삼성전자 주가 데이터 불러오기
start_date = "20230101"
end_date = "20240401"
ticker = "005930"

df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)

# 데이터프레임의 인덱스를 날짜 형식으로 변환
df.index = pd.to_datetime(df.index)

# ARIMA 모델을 사용한 주가 예측
# 종가 데이터를 사용
close_prices = df['종가']

# ARIMA 모델 학습 (p, d, q 값 설정 - 여기서는 (5, 1, 0)을 사용)
model = ARIMA(close_prices, order=(5, 1, 0))
model_fit = model.fit()

# 내일의 주가 예측
forecast = model_fit.forecast(steps=1)
predicted_price = forecast.iloc[0]

print(f"내일의 예측 주가: {predicted_price:.2f} KRW")

# 예측 결과 시각화
plt.figure(figsize=(12, 6))
plt.plot(close_prices.index, close_prices, label='Actual Close Price', color='blue')
plt.axvline(x=close_prices.index[-1], color='gray', linestyle='--')
plt.scatter(close_prices.index[-1] + pd.Timedelta(days=1), predicted_price, color='red', label='Predicted Price')
plt.title('Samsung Electronics (005930.KS) Close Price Prediction')
plt.xlabel('Date')
plt.ylabel('Price (KRW)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
# ARIMA 모델을 학습합니다.
# 7일간의 주가를 예측합니다.
# 예측 결과를 시각화합니다
```

```python
import pandas as pd
from pykrx import stock
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# 삼성전자 주가 데이터 불러오기
start_date = "20230101"
end_date = "20240610"
ticker = "005930"

df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)

# 데이터프레임의 인덱스를 날짜 형식으로 변환
df.index = pd.to_datetime(df.index)

# ARIMA 모델을 사용한 주가 예측
# 종가 데이터를 사용
close_prices = df['종가']

# ARIMA 모델 학습 (p, d, q 값 설정 - 여기서는 (5, 1, 0)을 사용)
model = ARIMA(close_prices, order=(5, 1, 0))
model_fit = model.fit()

# 향후 1주일 (7일)의 주가 예측
forecast = model_fit.forecast(steps=7)
predicted_prices = forecast.values

# 예측 결과 출력
print("향후 1주일 주가 예측:")
for i, price in enumerate(predicted_prices, 1):
    print(f"Day {i}: {price:.2f} KRW")

# 예측 결과 시각화
plt.figure(figsize=(12, 6))
plt.plot(close_prices.index, close_prices, label='Actual Close Price', color='blue')
plt.axvline(x=close_prices.index[-1], color='gray', linestyle='--')
future_dates = [close_prices.index[-1] + pd.Timedelta(days=i) for i in range(1, 8)]
plt.plot(future_dates, predicted_prices, label='Predicted Prices', color='red', marker='o')
plt.title('Samsung Electronics (005930.KS) Close Price Prediction for Next 7 Days')
plt.xlabel('Date')
plt.ylabel('Price (KRW)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

## KOSPI

```python
# 삼성전자 주가 데이터를 불러와서 종가 데이터를 사용하여 ARIMA 모델로 예측합니다.
# KOSPI 지수 데이터를 불러와서 종가 데이터를 사용하여 ARIMA 모델로 예측합니다.
# 예측된 주가와 지수 데이터를 바탕으로 수익률을 계산합니다.
# 예측된 주가와 수익률을 시각화합니다.
```

```python
import pandas as pd
from pykrx import stock
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# 삼성전자 주가 데이터 불러오기
start_date = "20230101"
end_date = "20240610"
ticker = "005930"

df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)
df.index = pd.to_datetime(df.index)

# KOSPI 지수 데이터 불러오기
kospi_df = stock.get_index_ohlcv_by_date(start_date, end_date, "1001")
kospi_df.index = pd.to_datetime(kospi_df.index)

# 삼성전자 종가 데이터
close_prices = df['종가']

# ARIMA 모델 학습 및 예측 함수
def arima_forecast(series, steps, order=(5, 1, 0)):
    model = ARIMA(series, order=order)
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=steps)
    return forecast

# 삼성전자 주가 예측 (향후 7일)
predicted_prices = arima_forecast(close_prices, steps=7)

# KOSPI 지수 예측 (향후 7일)
kospi_close_prices = kospi_df['종가']
predicted_kospi_prices = arima_forecast(kospi_close_prices, steps=7)

# 수익률 계산
def calculate_returns(prices):
    returns = (prices.diff() / prices.shift(1)) * 100
    return returns

# 삼성전자 주가 예측 수익률
predicted_prices_with_last = pd.concat([pd.Series([close_prices.iloc[-1]]), predicted_prices])
samsung_returns = calculate_returns(predicted_prices_with_last)

# KOSPI 예측 수익률
predicted_kospi_with_last = pd.concat([pd.Series([kospi_close_prices.iloc[-1]]), predicted_kospi_prices])
kospi_returns = calculate_returns(predicted_kospi_with_last)

# 예측 결과 출력
print("향후 1주일 주가 및 수익률 예측:")
for i, (price, kospi_price, s_return, k_return) in enumerate(zip(predicted_prices, predicted_kospi_prices, samsung_returns[1:], kospi_returns[1:]), 1):
    print(f"Day {i}: Samsung Price: {price:.2f} KRW, KOSPI Price: {kospi_price:.2f} KRW, Samsung Return: {s_return:.2f}%, KOSPI Return: {k_return:.2f}%")

# 예측 결과 시각화
plt.figure(figsize=(12, 12))

# 삼성전자 주가 예측 시각화
plt.subplot(3, 1, 1)
plt.plot(close_prices.index, close_prices, label='Actual Samsung Close Price', color='blue')
plt.axvline(x=close_prices.index[-1], color='gray', linestyle='--')
future_dates = [close_prices.index[-1] + pd.Timedelta(days=i) for i in range(1, 8)]
plt.plot(future_dates, predicted_prices, label='Predicted Samsung Prices', color='red', marker='o')
plt.title('Samsung Electronics (005930.KS) Close Price Prediction for Next 7 Days')
plt.xlabel('Date')
plt.ylabel('Price (KRW)')
plt.legend(loc='upper left')
plt.grid(True)

# KOSPI 지수 예측 시각화
plt.subplot(3, 1, 2)
plt.plot(kospi_close_prices.index, kospi_close_prices, label='Actual KOSPI Close Price', color='green')
plt.axvline(x=kospi_close_prices.index[-1], color='gray', linestyle='--')
future_dates_kospi = [kospi_close_prices.index[-1] + pd.Timedelta(days=i) for i in range(1, 8)]
plt.plot(future_dates_kospi, predicted_kospi_prices, label='Predicted KOSPI Prices', color='orange', marker='o')
plt.title('KOSPI Index Close Price Prediction for Next 7 Days')
plt.xlabel('Date')
plt.ylabel('Price (KRW)')
plt.legend(loc='upper left')
plt.grid(True)

# 예측 수익률 시각화
plt.subplot(3, 1, 3)
plt.plot(future_dates, samsung_returns[1:], label='Predicted Samsung Returns', color='red', marker='o')
plt.plot(future_dates_kospi, kospi_returns[1:], label='Predicted KOSPI Returns', color='orange', marker='o')
plt.title('Predicted Returns for Next 7 Days')
plt.xlabel('Date')
plt.ylabel('Returns (%)')
plt.legend(loc='upper left')
plt.grid(True)

plt.tight_layout()
plt.show()
```
