# 삼성전자 EDA

```python
# 기초 통계량 (describe()):
# mean, std, min, max 등의 기본 통계 정보를 제공합니다.

# 산점도 (Scatter Plot):
# Close 가격과 Volume 간의 관계를 시각화합니다.
# 이 그래프는 가격이 오를 때 거래량이 어떻게 변화하는지를 보여줄 수 있습니다.
```

```python
import FinanceDataReader as fdr
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 삼성전자 주가 데이터 가져오기 (2020년 1월 1일부터 현재까지)
df = fdr.DataReader('005930', '2020-01-01')

# 2. 인덱스를 리셋하여 'Date' 열로 변환
df = df.reset_index()

# 3. 기초 통계량 확인
basic_statistics = df.describe()
print("기초 통계량:")
print(basic_statistics)

# 4. 산점도 (Scatter Plot): 'Close'와 'Volume' 간의 관계
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Close', y='Volume', data=df)
plt.title('Scatter Plot of Close Price vs Volume')
plt.xlabel('Close Price (KRW)')
plt.ylabel('Volume')
plt.grid()
plt.show()
```

```python
# 데이터 수집:
# FinanceDataReader 라이브러리를 사용해 삼성전자(종목 코드: 005930)의 주가 데이터를 불러옵니다.

# 기본 EDA:
# info()와 describe()를 통해 데이터의 기본 정보와 기초 통계량을 확인합니다.
# 결측치를 확인하고, 필요한 경우 처리할 수 있습니다.

# 시각화:
# 시계열 그래프: 종가(Close)의 변화를 시간에 따라 시각화합니다.

# 산점도: 종가와 거래량(Volume) 간의 관계를 확인합니다.

# 히스토그램: 종가의 분포를 시각화하여 데이터의 분포 형태를 파악합니다.

# 상관 행렬 (Heatmap): 주가 데이터의 변수들 간 상관 관계를 시각화합니다.
```

```python
import FinanceDataReader as fdr
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 데이터 수집: 삼성전자 주가 데이터 (2020년 1월 1일부터 현재까지)
df = fdr.DataReader('005930', '2020-01-01')

# 2. 데이터 확인 및 기본 EDA
print("데이터의 기본 정보:")
print(df.info())
print("\n기초 통계량:")
print(df.describe())

# 3. 결측치 확인
print("\n결측치 확인:")
print(df.isnull().sum())

# 4. 시각화 1: 종가(Close) 시계열 그래프
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], label='Close Price', color='blue')
plt.title('Samsung Electronics Stock Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price (KRW)')
plt.legend()
plt.grid()
plt.show()

# 5. 시각화 2: 종가와 거래량의 산점도
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Close', y='Volume', data=df)
plt.title('Scatter Plot of Close Price vs Volume')
plt.xlabel('Close Price (KRW)')
plt.ylabel('Volume')
plt.grid()
plt.show()

# 6. 시각화 3: 종가 분포 (히스토그램)
plt.figure(figsize=(10, 6))
sns.histplot(df['Close'], bins=30, kde=True, color='purple')
plt.title('Distribution of Close Price')
plt.xlabel('Close Price (KRW)')
plt.ylabel('Frequency')
plt.grid()
plt.show()

# 7. 상관 분석
correlation = df[['Open', 'High', 'Low', 'Close', 'Volume']].corr()
print("\n상관계수:")
print(correlation)

# 8. 시각화 4: 상관 행렬 (Heatmap)
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Samsung Electronics Stock Data')
plt.show()
```

# 재무제표 EDA

```python
# 예제 데이터를 사용하여 기초 통계량과 시각화를 수행합니다.
# describe() 함수로 기초 통계량을 출력합니다.
# scatterplot()을 사용해 매출액과 영업이익 간의 관계를 시각화합니다.
# 연도별 매출액의 변화를 선 그래프로 표시합니다.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 예제 재무 데이터 생성
data = {
    'Year': ['2018', '2019', '2020', '2021', '2022'],
    'Revenue': [200000, 220000, 210000, 250000, 240000],
    'OperatingIncome': [30000, 35000, 33000, 40000, 38000],
    'NetIncome': [25000, 30000, 28000, 35000, 34000]
}
df = pd.DataFrame(data)

# 2. 기초 통계량 출력
print("기초 통계량:")
print(df.describe())

# 3. 매출액, 영업이익, 순이익 간의 관계 시각화
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Revenue', y='OperatingIncome', data=df)
plt.title('Scatter Plot of Revenue vs Operating Income')
plt.xlabel('Revenue (in Billion KRW)')
plt.ylabel('Operating Income (in Billion KRW)')
plt.grid()
plt.show()

# 4. 매출액의 연도별 변화 추이 시각화
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Revenue'], marker='o', label='Revenue')
plt.title('Annual Revenue Trend')
plt.xlabel('Year')
plt.ylabel('Revenue (in Billion KRW)')
plt.legend()
plt.grid()
plt.show()
```
