PCA(주성분 분석, Principal Component Analysis)는 고차원 데이터의 차원을 축소하여 데이터의 주요 특징을 추출하는 통계적 기법입니다. 

 이는 데이터에 내재된 복잡한 구조를 간단하게 나타내기 위해 사용되며, 데이터의 분산이 가장 큰 방향을 찾는 방식으로 작동합니다.

 큰 분산을 가지는 방향은 데이터 포인트들이 널리 퍼져 있는 방향으로, 해당 방향에는 데이터의 중요한 변화 정보가 많이 포함되어 있다고 볼 수 있습니다. 즉, 데이터의 분산이 크다는 것은 변동성이 크고 데이터의 중요한 특징을 많이 포함하고 있다는 의미입니다. 따라서, PCA에서는 분산이 큰 방향을 선택함으로써 데이터의 중요한 특징을 잃지 않고 최대한 잘 표현하고자 합니다.
    
PCA는 데이터 분석, 차원 축소, 데이터 시각화 등에 널리 사용되며, 데이터의 변형 및 압축에 매우 유용합니다.

코드 설명

```python
# PCA 객체 생성 (PCA(n_components=2)):
# 여기서는 n_components=2로 설정하여, 원래의 4차원 데이터를 두 개의 주성분(축)으로 축소합니다. 즉, 차원을 4에서 2로 줄입니다.
# 이 두 개의 주성분은 원본 데이터에서 분산을 가장 잘 설명하는 방향을 나타내며, 중요한 특성을 최대한 유지하면서 차원을 축소합니다.

# 데이터 변환 (pca.fit_transform(X)):
# fit_transform() 메소드를 사용하여 X 데이터를 2차원으로 변환합니다.
# fit() 단계에서 주성분을 찾고, transform() 단계에서 원래 데이터를 찾은 주성분 공간으로 투영합니다.
# 결과로 X_pca는 2차원으로 축소된 데이터를 담고 있습니다.
```

```python
# PCA 주성분이 데이터의 분산을 얼마나 설명하는지 출력

# explained_variance_ratio = pca.explained_variance_ratio_:
# 이 코드는 PCA 객체에서 주성분이 설명하는 분산의 비율을 가져옵니다.
# explained_variance_ratio_ 속성은 각 주성분이 원래 데이터의 분산에서 차지하는 비율을 나타내는 배열입니다.
# 예를 들어, [0.9246, 0.0531]이라면 첫 번째 주성분이 전체 분산의 약 92.46%, 두 번째 주성분이 약 5.31%를 설명하는 것입니다.
# print(f"Explained variance ratio by each principal component: {explained_variance_ratio}"):
# 각 주성분이 설명하는 분산의 비율을 출력합니다. 이 정보를 통해 각 주성분이 데이터의 중요한 변동을 얼마나 설명하는지 알 수 있습니다.
# print(f"Total explained variance: {explained_variance_ratio.sum()}"):
# 총 설명된 분산의 비율을 출력합니다.
# explained_variance_ratio.sum()은 두 주성분이 원래 데이터의 전체 분산 중 몇 퍼센트를 설명하는지를 나타냅니다.
# 총 설명된 분산 비율이 높을수록 축소된 차원이 원래 데이터의 정보를 잘 유지하고 있음을 의미합니다.
```

```python
# plt.figure(figsize=(8, 6)):
# 새로운 그림을 생성하고 크기를 설정합니다.
# figsize=(8, 6)은 그래프의 가로 크기를 8인치, 세로 크기를 6인치로 설정한 것입니다.
# 이 설정은 그래프의 크기를 조절하여 시각적으로 더 보기 좋게 만듭니다.

# colors = ['r', 'g', 'b']:
# 각 클래스(종류)에 대해 사용할 색상을 지정합니다.
# 'r', 'g', 'b'는 각각 빨간색 (red), 녹색 (green), **파란색 (blue)**을 나타내며, 데이터의 세 개의 클래스(붓꽃의 세 품종)에 대해 색상으로 구분해 시각화하는 데 사용됩니다.

# labels = iris.target_names:
# iris.target_names는 Iris 데이터셋에 있는 세 품종의 이름을 제공합니다: 'setosa', 'versicolor', 'virginica'.
# 이 labels는 그래프의 범례(legend)에 표시되어 각 색상이 어떤 품종을 나타내는지 알 수 있게 합니다.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Iris 데이터셋 로드
iris = load_iris()
X = iris.data  # 꽃받침 길이, 너비, 꽃잎 길이, 너비 (4차원 특성)
y = iris.target

# PCA 객체 생성 (2개의 주성분으로 축소)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)  # 4차원 데이터를 2차원으로 축소

# 설명된 분산 비율 출력
explained_variance_ratio = pca.explained_variance_ratio_
print(f"Explained variance ratio by each principal component: {explained_variance_ratio}")
print(f"Total explained variance: {explained_variance_ratio.sum()}")

# 결과 시각화
plt.figure(figsize=(8, 6))
colors = ['r', 'g', 'b']
labels = iris.target_names

# 각 클래스에 대해 다른 색상으로 표시
for i, color, label in zip(np.unique(y), colors, labels):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], color=color, label=label)

plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA of Iris Dataset')
plt.legend()
plt.grid()
plt.show()
```

첫 번째 주성분은 전체 분산의 약 **92.46%**를 설명합니다.
두 번째 주성분은 전체 분산의 약 **5.31%**를 설명합니다.
총 **97.77%**의 분산이 두 주성분을 통해 설명됩니다. 이는 대부분의 정보가 두 주성분에 의해 잘 설명된다는 의미입니다. 
따라서 원래의 4차원 데이터를 2차원으로 축소했음에도 불구하고, 데이터의 중요한 특성을 잘 유지하고 있음을 알 수 있습니다.

```python
# Principal Component Analysis (PCA) example in Python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Generating a sample dataset
np.random.seed(42)
num_samples = 100
num_features = 5

data = np.random.rand(num_samples, num_features)
columns = [f'Feature_{i+1}' for i in range(num_features)]
df = pd.DataFrame(data, columns=columns)

# Standardizing the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Applying PCA
n_components = 2
pca = PCA(n_components=n_components)
principal_components = pca.fit_transform(scaled_data)

# Creating a DataFrame for the principal components
pca_df = pd.DataFrame(data=principal_components, columns=[f'PC_{i+1}' for i in range(n_components)])

# Visualizing the results
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PC_1'], pca_df['PC_2'], c='b', marker='o')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Sample Dataset')
plt.grid()
plt.show()

# Explained variance ratio
explained_variance = pca.explained_variance_ratio_
print(f'Explained variance ratio of the components: {explained_variance}')
```

```python
# 첫 번째 주성분 (PC_1)은 전체 데이터의 분산의 약 **29.0%**를 설명합니다.
# 두 번째 주성분 (PC_2)은 전체 데이터의 분산의 약 **23.2%**를 설명합니다.
# 따라서 이 두 주성분은 원래 데이터의 분산 중 약 **52.2%**를 설명하고 있다고 해석할 수 있습니다. 이는 다섯 개의 원래 특성 중 두 개의 주요 축으로 데이터를 요약할 수 있으며, 그럼에도 데이터의 특성을 어느 정도 보존하고 있음을 의미합니다. PCA의 목적은 주로 데이터 차원을 줄여 복잡성을 낮추면서도 중요한 정보는 최대한 유지하는 것이므로, 이 정도의 분산 설명력이라면 적절한 축소라 할 수 있습니다.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 신용카드 사기 데이터 로드
# 데이터 파일이 로컬에 있다고 가정하고 'creditcard.csv'를 로드합니다.
data = pd.read_csv('creditcard.csv')

# 특성과 레이블 분리
X = data.drop(columns=['Class'])  # 'Class'는 레이블로 사용
y = data['Class']

# 데이터 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA 객체 생성 (2개의 주성분으로 축소)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)  # 원본 데이터를 2차원으로 축소

# 설명된 분산 비율 출력
explained_variance_ratio = pca.explained_variance_ratio_
print(f"Explained variance ratio by each principal component: {explained_variance_ratio}")
print(f"Total explained variance: {explained_variance_ratio.sum()}")

# 결과 시각화
plt.figure(figsize=(8, 6))
colors = ['r', 'b']
labels = ['Not Fraud', 'Fraud']

# 각 클래스에 대해 다른 색상으로 표시
for i, color, label in zip([0, 1], colors, labels):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], color=color, label=label, alpha=0.5)

plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA of Credit Card Fraud Dataset')
plt.legend()
plt.grid()
plt.show()
```

Total explained variance: 12.1%
