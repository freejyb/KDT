```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as pdr
import FinanceDataReader as fdr
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.graphics.tsaplots import plot_acf

# 주가 데이터 불러오기
samsung = fdr.DataReader('005930', start = '2015-01-01', end = '2023-04-03')

# 종가 데이터 추출하기
close_price = samsung['Close']


# AR(1) 모델 적합
model = AutoReg(close_price, lags=1)
results = model.fit()

# 예측값 구하기
start = len(close_price)
end = start + 5
predictions = results.predict(start=start, end=end, dynamic=False)
close_price = close_price.reset_index()

# 예측값 시각화
plt.plot(close_price.index, close_price["Close"], label='Observed')
plt.plot(predictions.index, predictions, label='Predicted')
plt.legend()
plt.show()

# ACF 그래프 시각화
plot_acf(close_price["Close"], lags=20)
plt.show()
print(close_price)
print(predictions)
```
