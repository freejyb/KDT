```python
# 백색 잡음의 특징
# 평균값이 0: 백색 잡음의 평균값은 보통 0입니다. 
# 따라서 시간이 지나면서 신호의 평균적인 변위가 없습니다.

# 모든 주파수에서 동일한 파워 스펙트럼 밀도: 백색 잡음은 각 주파수 대역에서 동일한 세기를 가지므로, 주파수 스펙트럼 상에서 평탄한 선으로 나타납니다. 이는 신호가 특정 주파수에 편향되지 않고 모든 주파수가 동일한 세기로 존재한다는 의미입니다.
# 시간 상관성이 없음: 백색 잡음의 신호값은 서로 독립적이고, 서로 상관성이 없습니다. 
# 이는 특정 시간에 발생한 신호값이 다른 시간의 신호값에 영향을 미치지 않는다는 의미입니다.
```

```python
# 백색 잡음 생성
duration = 1.0  # 신호의 길이 (초)
sampling_rate = 1000  # 샘플링 주파수 (Hz)
num_samples = int(duration * sampling_rate)  # 총 샘플 개수
```

```python
# 백색 잡음 생성
# **duration**은 신호의 전체 길이를 초 단위로 설정하는 변수입니다. 여기서는 신호가 1초 동안 지속된다고 가정하고, duration을 1.0으로 설정했습니다.
# 이 값에 따라 전체 샘플 수가 결정됩니다.

# sampling_rate = 1000  # 샘플링 주파수 (Hz)
# **sampling_rate**는 샘플링 주파수로, 1초 동안 샘플을 몇 번 측정할 것인지 결정하는 변수입니다.
# 여기서는 1000 Hz로 설정했으므로, 1초 동안 1000개의 샘플을 생성합니다.
# 이 값이 높을수록 신호가 더 세밀하게 샘플링됩니다.

# num_samples = int(duration * sampling_rate)  # 총 샘플 개수
# **num_samples**는 전체 신호에서 필요한 샘플의 총 개수를 계산하는 변수입니다.
# **duration * sampling_rate**를 계산하여 duration 동안 1000Hz의 속도로 샘플링했을 때의 총 샘플 개수를 구합니다.
# 예를 들어, duration이 1초이고 sampling_rate가 1000Hz일 경우, 총 샘플 수는 1000개가 됩니다.
```

```python
# 백색잡음 신호생성

# np.random.normal(0, 1, num_samples):
# np.random.normal은 정규 분포(가우시안 분포)를 따르는 난수를 생성하는 함수입니다.
# **첫 번째 매개변수 0**은 평균(mean)을 설정합니다. 이 경우 백색 잡음의 평균은 0으로 설정됩니다.
# **두 번째 매개변수 1**은 표준편차(standard deviation)를 설정합니다. 여기서는 표준편차가 1인 정규 분포를 따르는 백색 잡음을 생성합니다.
# **세 번째 매개변수 num_samples**은 생성할 샘플의 개수로, duration과 sampling_rate를 곱하여 계산한 총 샘플 개수입니다. 이만큼의 데이터 포인트를 가진 백색 잡음 신호가 생성됩니다.
# 결과적으로 white_noise는 길이가 num_samples인 배열로, 각 요소는 평균 0과 표준편차 1의 정규 분포를 따르는 값입니다. 이 배열이 백색 잡음 신호를 나타냅니다.
```

```python
# 백색 잡음 신호 생성 (평균 0, 표준편차 1의 가우시안 분포)
white_noise = np.random.normal(0, 1, num_samples)
```

```python
# # 시간 축 생성
# np.linspace(0, duration, num_samples):

# np.linspace는 지정된 구간 내에서 일정한 간격으로 값을 생성하는 NumPy 함수입니다.
# 이 함수를 사용하여 신호의 시작 시점부터 종료 시점까지 일정한 간격으로 시간값을 생성합니다.
# 매개변수 설명:
# 첫 번째 매개변수 0: 시간 축의 시작점을 설정합니다. 여기서는 0초에서 시작합니다.
# 두 번째 매개변수 duration: 시간 축의 끝점을 설정합니다. 이 값은 신호의 길이(초)로, duration만큼의 시간이 흐르는 동안의 시간값을 생성합니다. 예를 들어, duration이 1.0초라면 시간 축은 0초에서 1초까지 생성됩니다.
# 세 번째 매개변수 num_samples: 생성할 값의 개수를 지정합니다. 신호의 각 샘플마다 대응하는 시간값이 필요하기 때문에, 전체 샘플 개수와 동일한 개수의 시간값을 생성합니다.

# 결과:
# time 변수는 0초부터 duration까지의 시간값을 가지는 배열로, 길이는 num_samples입니다. 각 값은 신호 샘플에 대한 시간 정보를 나타냅니다.
# 예를 들어, duration이 1.0초이고 sampling_rate가 1000Hz라면 time 배열에는 0초에서 1초까지의 시간값이 1000개 생성됩니다.
```

```python
# 시간 축 생성
time = np.linspace(0, duration, num_samples)
```

```python
# # 백색 잡음 시각화
# plt.figure(figsize=(10, 4))는 새로운 그림을 생성하고, 그림의 크기를 설정합니다.
# **figsize=(10, 4)**는 그림의 가로 크기와 세로 크기를 각각 10과 4로 설정합니다. 가로로 긴 형태의 그래프가 만들어지며, 그래프가 더 보기 쉽게 표시됩니다.

# plt.plot(time, white_noise, color="gray")
# plt.plot는 time 배열을 x축으로, white_noise 배열을 y축으로 설정하여 그래프를 그립니다.
# time: 각 샘플의 시간 정보를 나타내며, x축 값입니다.
# white_noise: 백색 잡음 신호의 값으로, y축 값입니다.
# color="gray": 그래프의 선 색깔을 회색으로 지정하여 잡음 신호의 무작위성을 강조합니다.

# plt.title("White Noise Signal")
# plt.title는 그래프의 제목을 설정합니다.
# **"White Noise Signal"**이라는 제목을 그래프 상단에 표시하여, 그래프가 백색 잡음을 나타낸다는 것을 명확히 합니다.

# plt.xlabel("Time (s)")
# plt.xlabel은 x축의 라벨(제목)을 설정합니다.
# **"Time (s)"**는 시간 단위를 초(second)로 표시하여, x축이 시간을 나타낸다는 것을 보여줍니다.

# plt.ylabel("Amplitude")
# plt.ylabel은 y축의 라벨(제목)을 설정합니다.
# **"Amplitude"**는 신호의 진폭을 의미하며, y축 값이 백색 잡음의 진폭을 나타낸다는 것을 알려줍니다.

# plt.show()
# plt.show는 그래프를 화면에 표시합니다.
# 이 명령을 통해 생성된 백색 잡음 신호 그래프가 시각적으로 출력됩니다.
```

```python
# 백색 잡음 시각화
plt.figure(figsize=(10, 4))
plt.plot(time, white_noise, color="gray")
plt.title("White Noise Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 백색 잡음 생성
duration = 1.0  # 신호의 길이 (초)
sampling_rate = 1000  # 샘플링 주파수 (Hz)
num_samples = int(duration * sampling_rate)  # 총 샘플 개수

# 백색 잡음 신호 생성 (평균 0, 표준편차 1의 가우시안 분포)
white_noise = np.random.normal(0, 1, num_samples)

# 시간 축 생성
time = np.linspace(0, duration, num_samples)

# 백색 잡음 시각화
plt.figure(figsize=(10, 4))
plt.plot(time, white_noise, color="gray")
plt.title("White Noise Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

## 확률보행과정 (Random Walk Process)

```python
# Probability to move up or down
# up = 40%, down= 60%
prob = [0.4, 0.6]  
  
# starting position
start = 10000
rand_walks = [start]

# creating the random points
rr = np.random.random(250)
downp = rr < prob[0]
upp = rr >= prob[1]

# random walk process
# z(t) = z(t-1) + a(t), where a(t) is white noise
for idown, iup in zip(downp, upp):
    rand_walks.append(rand_walks[-1] - idown + iup)

# plot 
plt.plot(rand_walks)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as pdr
import FinanceDataReader as fdr
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.graphics.tsaplots import plot_acf

# 주가 데이터 불러오기
samsung = fdr.DataReader('005930', start = '2015-01-01', end = '2023-04-03')

# 종가 데이터 추출하기
close_price = samsung['Close']


# AR(1) 모델 적합
model = AutoReg(close_price, lags=1)
results = model.fit()

# 예측값 구하기
start = len(close_price)
end = start + 5
predictions = results.predict(start=start, end=end, dynamic=False)
close_price = close_price.reset_index()

# 예측값 시각화
plt.plot(close_price.index, close_price["Close"], label='Observed')
plt.plot(predictions.index, predictions, label='Predicted')
plt.legend()
plt.show()

# ACF 그래프 시각화
plot_acf(close_price["Close"], lags=20)
plt.show()
print(close_price)
print(predictions)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as pdr
import FinanceDataReader as fdr
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.graphics.tsaplots import plot_acf

# 주가 데이터 불러오기
samsung = fdr.DataReader('005930', start = '2015-01-01', end = '2023-04-05')

# 종가 데이터 추출하기
close_price = samsung['Close']


# AR(1) 모델 적합
model = AutoReg(close_price, lags=10)
results = model.fit()

# 예측값 구하기
start = len(close_price)
end = start + 5
predictions = results.predict(start=start, end=end, dynamic=False)
close_price = close_price.reset_index()

# 예측값 시각화
plt.plot(close_price.index, close_price["Close"], label='Observed')
plt.plot(predictions.index, predictions, label='Predicted')
plt.legend()
plt.show()

# ACF 그래프 시각화
plot_acf(close_price["Close"], lags=20)
plt.show()
print(close_price)
print(predictions)
```
